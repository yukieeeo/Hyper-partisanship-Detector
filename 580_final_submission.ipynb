>>Motivation

   Accuracy, fairness and timeliness are the three main measures to evaluate news. Accuracy comes from constant attention and repetitive details check of the news. Timeliness comes from increasing knowledge, confidence and experience of the news reporters. Fairness comes from objectivity and impartiality of the news [1]. However, media today lacks strict codes of conduct; such reporting will mislead readers. The credibility of the news agencies will also be infringed. One of the most frequent biased news reporting is on hyperpartisanship meaning the news is extremely in favor of one political party. Publishers can be biased to one political party on various purposes. Some readers are able to recognize whether the news and reports are partisan, but some are not. Therefore, the political affiliation of publishers are essential when assessing their news quality. Publishers can take extreme left-wing or right-wing standpoint since not all news are carefully examined by the committee of editors. These views affect readers opinions in various ways. Thus, in order to increase the awareness of fair and objective news reporting, it is meaningful to study the partisanship of publishers. Moreover, if hyperpartisanship can be automatically detected by the content of the articles, then it is much easier for researchers to determine the hyperpartisan-level of the publisher and make instructive recommendations for them. Above all, the main question for the team's study is whether a specific media has hyper-partisanship? The specific media company the team chose is CNN news media. 

>>Relevant background research

   This project is mainly based on a competition report UBC-NLP at SemEval-2019 Task 4: Hyper partisan News Detection With Attention-Based Bi-LSTMs[2]. Instead of only using traditional classifiers such as Naive Bayes and logistic regression classifiers, the report applies deep learning models to detect hyper-partisan articles and acquires outstanding results. Concretely speaking, the researchers employed 5 different models: RNN, CNN, LSTM, CNN and LSTM combination, and RNN and LSTM combination. These 5 models performances were evaluated, and they concluded that the combination of CNN and LSTM achieves the best performance at the end. Based on their work, the team used KNN, logistic regression classifier, random forest, SVM, Nave Bayes, CNN and LSTM models to develop the hyper partisan news detector. The team was unable to perform CNN and LSTM combination models because it was very time consuming to run the model.

Data (3) size is over 2 GB. Besides, the team web scrapped 50 articles titles from CNN news network.

Recall measures the percentage of the total relevant datapoints in a database retrieved by the model. Precision measures the percentage of relevant datapoints in relation to the number of datapoints retrieved. The CNN model has the highest accuracy which is 91.06%. The CNN model also has the highest recall and precision so this model is the most optimal model.
```python
# define the function to clean data
eng_stopwords = set(stopwords.words('english'))

def clean_text(text, remove_stopwords=False):
    text = re.sub(r'https?:\/\/.+$', ' ',text)
    text = re.sub(r'[^a-zA-Z]', ' ', text)
    text = re.sub(r'\b\w\b', ' ', text)
    text = re.sub(r'\bth\b', ' ', text)
    text = re.sub(r'\bst\b', ' ', text)
    text = re.sub(r'\bone\b', ' ', text)
    text = re.sub(r'\btwo\b', ' ', text)
    text = re.sub(r'\bthree\b', ' ', text)
    text = re.sub(r'\bfive\b', ' ', text)
    words = text.lower().split()
    if remove_stopwords:
        words = [w for w in words if w not in eng_stopwords]
    return words

# define a function to transfer title to vector
def to_title_vector(review):
    global word_vec
    
    review = clean_text(review, remove_stopwords=True)
    word_vec = np.zeros((1,300))
    
    for word in review:
        if word in model:
            word_vec += np.array([model[word]])
    return pd.Series(word_vec.mean(axis = 0))

```

```python
# Define a function for plot confusion matrix
import matplotlib.pyplot as plt
import itertools
def plot_confusion_matrix(cm, classes,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    """
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=0)
    plt.yticks(tick_marks, classes)

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
# In deep learning models, we transfer each title to a matrix (300*30)
# For every title, we only use first 30 words
# every word in the title can be transfer to a vecter(300)

max_word_length=30
for i in range(len(df)//3000+1):
    train_1 = np.zeros((max_word_length,300))
    train_1=train_1[np.newaxis]
    for title in df.title[i*3000:min((i+1)*3000,len(df))]:
        title = clean_text(title, remove_stopwords=True)
        word_vec = np.zeros((1,300))
        count=0
        for word in title: 
            if word in model and count<max_word_length:
                word_vec=np.append(word_vec,np.array([model[word]]),axis=0)
                count=count+1
        word_vec=np.delete(word_vec,0,0)
        
        while len(word_vec) < max_word_length:
            word_vec=np.append(word_vec,np.zeros((1,300)),axis=0)
        word_vec=word_vec[np.newaxis]
        train_1=np.append(train_1,word_vec,axis=0)
    train_1=np.delete(train_1,0,0)
    np.save('train_%d.npy'%(i+1),train_1)
    print('get%d'%(i+1))
```

# LSTM

### Prepare the training and test dataset for LSTM 

```python
import tensorflow as tf
from tensorflow.contrib import rnn
sess=tf.Session
```

```python
train_x_1=np.load('train_x_1.npy')
```

```python
hyper_list=[]
for i in df.label:
    if i==True:
        hyper_list.append([1, 0])  
    if i==False:
        hyper_list.append([0, 1])
```

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(train_x_1,hyper_list,test_size = 0.2, random_state = 0)
del train_x_1
```

### Build the LSTM

```python
lr=1e-3
input_size=300   # every word has 300 features
timestep_size=30  #only use first 50 words in the title
hidden_size=1000   # the number of neurons in the hidden layer is 1000
layer_num=2        # 2 hidden layer
class_num=2
batch_size=50
num_batch = len(y_train) // batch_size

X = tf.placeholder(tf.float32, [None, 30,300])
y = tf.placeholder(tf.float32)
keep_prob = tf.placeholder(tf.float32)
```

```python
def lstm_cell():
    cell=rnn.LSTMCell(hidden_size,reuse=tf.get_variable_scope().reuse)
    return rnn.DropoutWrapper(cell,output_keep_prob=keep_prob)

mlstm_cell=tf.contrib.rnn.MultiRNNCell([lstm_cell() for _ in range(layer_num)],state_is_tuple=True)

init_state=mlstm_cell.zero_state(batch_size,dtype=tf.float32)

outputs=list()
state=init_state
with tf.variable_scope('RNN') :
    for timestep in range(timestep_size):
        if timestep>0:
            tf.get_variable_scope().reuse_variables()       
        (cell_output,state)=mlstm_cell(X[:,timestep,:],state)
        outputs.append(cell_output)
h_state=outputs[-1]
```

```python
W=tf.Variable(tf.truncated_normal([hidden_size,class_num],stddev=0.1),dtype=tf.float32)
bias=tf.Variable(tf.constant(0.1,shape=[class_num]),dtype=tf.float32)
y_pre=tf.nn.softmax(tf.matmul(h_state,W)+bias)

cross_entropy=-tf.reduce_mean(y*tf.log(y_pre))
train_op=tf.train.AdamOptimizer(lr).minimize(cross_entropy)

correct_prediction = tf.equal(tf.argmax(y_pre,1), tf.argmax(y,1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction, "float"))

TP = tf.count_nonzero(tf.argmax(y_pre,1)* tf.argmax(y,1))
TN = tf.count_nonzero((tf.argmax(y_pre,1)- 1) * (tf.argmax(y,1) - 1))
FP = tf.count_nonzero(tf.argmax(y_pre,1) * (tf.argmax(y,1) - 1))
FN = tf.count_nonzero((tf.argmax(y_pre,1) - 1) * tf.argmax(y,1))
precision = TP / (TP + FP)
recall = TP / (TP + FN)
f1 = 2 * precision * recall / (precision + recall)
```

### Train the LSTM

```python
sess = tf.Session()
init = tf.global_variables_initializer()
sess.run(init)

for e in range(200):
    for j in range(num_batch):#
        batch_X = X_train[j*batch_size : (j+1)*batch_size]
        batch_y = y_train[j*batch_size : (j+1)*batch_size]
        _= sess.run(train_op, feed_dict={X:batch_X, y:batch_y, keep_prob:0.5})

    if e % 2 == 0:
        print('epoch:',e,'iter:', j)
        F1_TRAIN = f1.eval(session=sess, feed_dict = {X:batch_X , y: batch_y,keep_prob:0.5})
        print("F1_train: {}".format(F1_TRAIN))
        recall_TRAIN = recall.eval(session=sess, feed_dict = {X:batch_X , y: batch_y,keep_prob:0.5})
        print("recall_train: {}".format(recall_TRAIN))
        trainAccuracy = accuracy.eval(session=sess, feed_dict={X:batch_X , y: batch_y,keep_prob:0.5})
        print ("step %d, training accuracy %g"%(e, trainAccuracy))
        acc = accuracy.eval(session=sess, feed_dict = {X: X_test[0:batch_size], y: y_test[0:batch_size],keep_prob:0.5})
        print("testing accuracy: {}".format(acc))
        
        tp = TP.eval(session=sess,  feed_dict = {X: X_test[0:batch_size], y: y_test[0:batch_size],keep_prob:0.5})
        print("TP: {}".format(tp))
        tn = TN.eval(session=sess,  feed_dict = {X: X_test[0:batch_size], y: y_test[0:batch_size],keep_prob:0.5})
        print("TN: {}".format(tn))  
        fp = FP.eval(session=sess,  feed_dict = {X: X_test[0:batch_size], y: y_test[0:batch_size],keep_prob:0.5})
        print("FP: {}".format(fp))          
        fn = FN.eval(session=sess,  feed_dict = {X: X_test[0:batch_size], y: y_test[0:batch_size],keep_prob:0.5})
        print("FN: {}".format(fn)) 
        F1 = f1.eval(session=sess,  feed_dict = {X: X_test[0:batch_size], y: y_test[0:batch_size],keep_prob:0.5})
        print("F1: {}".format(F1))
        
        RECALL = recall.eval(session=sess, feed_dict = {X: X_test[0:batch_size], y: y_test[0:batch_size],keep_prob:0.5})
        print("recall: {}".format(RECALL))
        Precision = precision.eval(session=sess,  feed_dict = {X: X_test[0:batch_size], y: y_test[0:batch_size],keep_prob:0.5})
        print("precision: {}".format(Precision))
        
        print('\n')        
```

### Prepare the training and test dataset for CNN

```python
# load sentence matrix
import pandas as pd
import numpy as np
train_x_1=np.load('train_x_1.npy')
train_x_2=np.expand_dims(train_x_1,-1)
del train_x_1

# make the label 
df = pd.read_csv('2018data.csv')
hyper_list=[]
for i in df.label:
    if i==True:
        hyper_list.append([1, 0])  
    if i==False:
        hyper_list.append([0, 1])
        
# split the training set and test set
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(train_x_2,hyper_list,test_size = 0.2, random_state = 0)
del train_x_2
```

### The structure of cnn is shown below: Yoon Kim-Convolutional Neural Networks for Sentence Classification (2014 EMNLP)

![title](1.png)

```python
# define cnn structure
import tensorflow as tf
tf.reset_default_graph()

def cnn(X,dropout_keep_prob):       
    input_size = 30
    num_classes = 2
    embedding_size=300
    num_filters=128
    filter_sizes = [3,4,5]
    pooled_outputs = []
    for i, filter_size in enumerate(filter_sizes):
        with tf.name_scope("conv-maxpool-%s" % filter_size):
            filter_shape = [filter_size, embedding_size, 1, num_filters]
            W = tf.Variable(tf.truncated_normal(filter_shape, stddev=0.1))
            b = tf.Variable(tf.constant(0.1, shape=[num_filters]))
            conv = tf.nn.conv2d(X, W, strides=[1, 1, 1, 1], padding="VALID")
            h = tf.nn.relu(tf.nn.bias_add(conv, b))
            pooled = tf.nn.max_pool(h, ksize=[1, input_size - filter_size + 1, 1, 1], strides=[1, 1, 1, 1], padding='VALID')
            pooled_outputs.append(pooled)
    num_filters_total = num_filters * len(filter_sizes)
    h_pool = tf.concat(pooled_outputs,3)
    h_pool_flat = tf.reshape(h_pool, [-1, num_filters_total])
    
    # dropout
    with tf.name_scope("dropout"):
        h_drop = tf.nn.dropout(h_pool_flat, dropout_keep_prob)
        
    with tf.name_scope("output"):
        W_fc = tf.get_variable("W_fc", shape=[num_filters_total, num_classes], initializer=tf.contrib.layers.xavier_initializer())
        b_fc = tf.Variable(tf.constant(0.1, shape=[num_classes]))
        output = tf.nn.xw_plus_b(h_drop, W_fc, b_fc)
    return output
```

```python
# train the model
def train_neural_network(output):
    batch_size = 64
    num_batch = len(y_train) // batch_size
    
    optimizer = tf.train.AdamOptimizer(1e-3)
    loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=output, labels=y))
    grads_and_vars = optimizer.compute_gradients(loss)
    train_op = optimizer.apply_gradients(grads_and_vars)
    correct_prediction = tf.equal(tf.argmax(output,1), tf.argmax(y,1))
    accuracy = tf.reduce_mean(tf.cast(correct_prediction, "float"))
    TP = tf.count_nonzero(tf.argmax(output,1)* tf.argmax(y,1))
    TN = tf.count_nonzero((tf.argmax(output,1)- 1) * (tf.argmax(y,1) - 1))
    FP = tf.count_nonzero(tf.argmax(output,1) * (tf.argmax(y,1) - 1))
    FN = tf.count_nonzero((tf.argmax(output,1) - 1) * tf.argmax(y,1))
    precision = TP / (TP + FP)
    recall = TP / (TP + FN)
    f1 = 2 * precision * recall / (precision + recall)
    summary_op=tf.summary.merge_all()

    saver = tf.train.Saver(tf.global_variables())
    with tf.Session() as sess:
        sess.run(tf.global_variables_initializer())
        
        for e in range(201):
            for j in range(num_batch):
                batch_X = X_train[j*batch_size : (j+1)*batch_size]
                batch_y = y_train[j*batch_size : (j+1)*batch_size]
                _, loss_ = sess.run([train_op, loss], feed_dict={X:batch_X, y:batch_y, dropout_keep_prob:0.5})
                
                if j % 1000 == 0:
                    print('epoch:',e,'iter:', i,'loss:', loss_)
                
            if e % 10 == 0:
                F1_TRAIN = f1.eval(session=sess, feed_dict = {X:batch_X , y: batch_y,dropout_keep_prob:0.5})
                print("F1_train: {}".format(F1_TRAIN))
                recall_TRAIN = recall.eval(session=sess, feed_dict = {X:batch_X , y: batch_y,dropout_keep_prob:0.5})
                print("recall_train: {}".format(recall_TRAIN))
                trainAccuracy = accuracy.eval(session=sess, feed_dict={X:batch_X , y: batch_y,dropout_keep_prob:0.5})
                print ("step %d, training accuracy :%g"%(e, trainAccuracy))
                acc = accuracy.eval(session=sess, feed_dict = {X: X_test, y: y_test,dropout_keep_prob:0.5})
                print("testing accuracy: {}".format(acc))
                tp = TP.eval(session=sess, feed_dict = {X: X_test, y: y_test,dropout_keep_prob:0.5})
                print("TP: {}".format(tp))
                tn = TN.eval(session=sess, feed_dict = {X: X_test, y: y_test,dropout_keep_prob:0.5})
                print("TN: {}".format(tn))  
                fp = FP.eval(session=sess, feed_dict = {X: X_test, y: y_test,dropout_keep_prob:0.5})
                print("FP: {}".format(fp))
                fn = FN.eval(session=sess, feed_dict = {X: X_test, y: y_test,dropout_keep_prob:0.5})
                print("FN: {}".format(fn)) 
                F1 = f1.eval(session=sess, feed_dict = {X: X_test, y: y_test,dropout_keep_prob:0.5})
                print("F1: {}".format(F1))
                RECALL = recall.eval(session=sess, feed_dict = {X: X_test, y: y_test,dropout_keep_prob:0.5})
                print("recall: {}".format(RECALL))
                Precision = precision.eval(session=sess, feed_dict = {X: X_test, y: y_test,dropout_keep_prob:0.5})
                print("precision: {}".format(Precision))
                saver.save(sess, "./model/hyperpartisan", global_step=e)
                print('\n')
                
tf.reset_default_graph()
X = tf.placeholder(tf.float32, [None, 30,300,1])
y = tf.placeholder(tf.float32)
dropout_keep_prob = tf.placeholder(tf.float32)
output=cnn(X,dropout_keep_prob)
train_neural_network(output)
```
